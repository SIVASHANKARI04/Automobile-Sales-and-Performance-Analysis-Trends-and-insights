##2.HOW CAN YOU CALCULATE THE AVERAGE SALES FOR EACH MANUFACTURER AND DISPLAY THE TOP 5 MANUFACURER BY AVERAGE SALES?


```python
df.Manufacturer.unique()
```

```python
df1=df.groupby("Manufacturer")["Sales_in_thousands"].mean()
```

```python
df2=df1.sort_values(ascending=False)
```
